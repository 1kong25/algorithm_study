```python
def least(a,b):
    A,B=a,b
    while b>0:
        a,b=b,a%b
    gcd=a # 
    return A*B // gcd #   

def solution(arr):
    arr.sort()
    answer=arr[0]
    for i in range(0, len(arr)-1):
        answer=least(answer,arr[i+1])
    return answer
```

```python
solution([2,6,8,14])
```

```python
def solution(name):
    answer = 0
    #     -1
    min_move = len(name) -1
    for i in range(len(name)):
        #     
        answer += min(ord(name[i])-ord("A"),ord("Z")+1-ord(name[i]))
        # ord():           - ord('a')=97
        # chr():      - chr(97)
    
        #     A  
        next = i + 1
        while next < len(name) and name[next] == 'A':
            next += 1
        
        # ,  A  ,  A     
        min_move = min([min_move, 2 *i + len(name) - next, i + 2 * (len(name) -next)])
    #  ()        
    answer += min_move
    
    return answer
```

```python
#             
def solution(s,n):
    lower='abcdefghijklmnopqrstuvwxyz'
    upper='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    answer=[]
    for i in s:
        if i==' ':
            answer.append(' ')
        elif i.islower() is True:
            new_=lower.find(i)+n #         find
            answer.append(lower[new_%26])
        else:
            new_=upper.find(i)+n
            answer.append(upper[new_%26])
    return ''.join(answer)
            


##    =>  .    
#    -       .
def solution(s, n):
    answer = ''
    for i in range(len(s)):
        if s[i]=='z' or s[i]=='Z':
            answer+=chr(ord(s[i])+n-(123-97))
        elif ord(s[i])==32:
            answer+=' '
        else:
            answer+=chr(ord(s[i])+n)
    return answer
```
