## 프로그래머스 level 2. 배열의 최소공배수 구하기


```python
def least(a,b):
    A,B=a,b
    while b>0:
        a,b=b,a%b
    gcd=a # 최대공약수
    return A*B // gcd # 두 수의 최소공배수

def solution(arr):
    arr.sort()
    answer=arr[0]
    for i in range(0, len(arr)-1):
        answer=least(answer,arr[i+1])
    return answer
```

```python
solution([2,6,8,14])
```

## 조이스틱으로 알파벳 이름 완성

```python
def solution(name):
    answer = 0
    # 기본 최소 좌우이동 횟수는 길이-1
    min_move = len(name) -1
    for i in range(len(name)):
        # 해당 알파벳 변경 최솟값 추가
        answer += min(ord(name[i])-ord("A"),ord("Z")+1-ord(name[i]))
        # ord(문자): 하나의 문자를 인자로 받고 해당 문자에 해당하는 아스키코드 정수를 반환 - ord('a')=97
        # chr(정수): 정수에 해당하는 아스키코드 문자 반환 - chr(97)
    
        # 해당 알파벳 다음부터 연속된 A 문자열 찾기
        next = i + 1
        while next < len(name) and name[next] == 'A':
            next += 1
        
        # 기존, 연속된 A의 왼쪽시작 방식, 연속된 A의 오른쪽시작 방식 비교 및 갱신
        min_move = min([min_move, 2 *i + len(name) - next, i + 2 * (len(name) -next)])
    # 알파벳 변경(상하이동) 횟수에 좌우이동 횟수 추가    
    answer += min_move
    
    return answer
```

## 문자열 내림차순 정렬

```python
# sorted는 문자열을 list 타입으로 바꿔서 정렬함. 
# 역순으로 정렬 후 join을 통해 리스트 값들을 다시 문자열로 반환
def solution(s):
    return ''.join(sorted(s, reverse=True))



## 내가 풀이한 방법 => 아스키코드로 바꿔서 리스트로 만들고 정렬한 후 다시 문자로 바꿈
# 완전 비효율적이었다....
def solution(s):
    answer = []
    string=''
    for i in range(len(s)):
        answer.append(ord(s[i]))
    answer.sort()
    answer.reverse()
    for i in range(len(s)):
        string+=chr(answer[i])
    return string
```

## 시저 암호
🤬 실패한 문제

```python
# 리스트 인덱스 번호를 알파벳 수로 나눈 나머지 이용해 새 인덱스 번호 알아내는 방법
def solution(s,n):
    lower='abcdefghijklmnopqrstuvwxyz'
    upper='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    answer=[]
    for i in s:
        if i==' ':
            answer.append(' ')
        elif i.islower() is True:
            new_=lower.find(i)+n # 찾는 값에 해당하는 값의 인덱스 번호 찾는 함수 find
            answer.append(lower[new_%26])
        else:
            new_=upper.find(i)+n
            answer.append(upper[new_%26])
    return ''.join(answer)
            


## 내가 풀이한 방법 => 복잡하긴 하다. 아스키코드 숫자 기준으로 생각함
# 테스트 통과 못함 - 테스트케이스는 다 맞았는데 왜 실패인지 잘 모르겠따.
def solution(s, n):
    answer = ''
    for i in range(len(s)):
        if s[i]=='z' or s[i]=='Z':
            answer+=chr(ord(s[i])+n-(123-97))
        elif ord(s[i])==32:
            answer+=' '
        else:
            answer+=chr(ord(s[i])+n)
    return answer
```
